```python
df = pd.read_csv("stores_sales_forecasting.csv", encoding="latin1")
df.head()
```

```python
df.columns
df.head()
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', dayfirst=True)
df = df.sort_values('Order Date')
```

```python
df['Order Date'].isna().sum()
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales = monthly_sales.to_timestamp()
```

```python
monthly_sales.head()
monthly_sales.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(monthly_sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
rolling_mean = monthly_sales.rolling(3).mean()

plt.figure(figsize=(12,5))
plt.plot(monthly_sales, label="Original")
plt.plot(rolling_mean, label="Rolling Mean (3M)")
plt.legend()
plt.show()
```

```python
train = monthly_sales[:-6]
test = monthly_sales[-6:]
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(
    train,
    trend='add',
    seasonal='add',
    seasonal_periods=12
)

fit_model = model.fit()
```

```python
forecast = fit_model.forecast(6)

plt.figure(figsize=(12,5))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.plot(forecast, label="Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

mae = mean_absolute_error(test, forecast)
mape = np.mean(np.abs((test - forecast) / test)) * 100

print("MAE:", mae)
print("MAPE:", mape)
```

```python
forecast_df = pd.DataFrame({
    "Date": forecast.index,
    "Forecasted_Sales": forecast.values
})

forecast_df.to_csv("forecast_output.csv", index=False)
```
